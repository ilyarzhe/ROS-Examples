```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_table('data/healthdata.txt')
```

```python
df['country'] = df['country "spending" "lifespan"'].str.split().str[0]
df['spending'] = df['country "spending" "lifespan"'].str.split().str[1].astype(float)
df['lifespan'] = df['country "spending" "lifespan"'].str.split().str[2].astype(float)
df.head()
```

```python
px.scatter(df, x='spending', y='lifespan', text='country',
           title='Health Expenditure vs Life Expectancy')
```
